## Feature Scaling
We need to scale only the `Amount` column as all other columns are already scaled by the PCA transformation.

```python
# Standardization method
from sklearn.preprocessing import StandardScaler
```

```python
# Instantiate the Scaler
scaler = StandardScaler()
```

```python
# Fit the data into scaler and transform
X['Amount'] = scaler.fit_transform(X[['Amount']])
```

```python
X.head()
```

# Power transform on skweness dataset

```python
# Listing the columns
cols = X.columns
cols
```

```python
# Importing PowerTransformer
from sklearn.preprocessing import PowerTransformer
# Instantiate the powertransformer
pt = PowerTransformer(method='yeo-johnson', standardize=True, copy=False)
# Fit and transform the PT on training data
X[cols] = pt.fit_transform(X)
```

```python
X.head()
```

# Handling data imbalanc

- SMOTE :- Synthetic minority oversampling technique. It is another oversampling technique, which uses nearest neighbor algorithm to create synthetic data.


## SMOTE (Synthetic Minority Oversampling Technique)

We are creating synthetic samples by doing upsampling using SMOTE(Synthetic Minority Oversampling Technique).

Imbalanced datasets pose a common challenge for machine learning practitioners in binary classification problems. This scenario frequently arises in practical business applications like fraud detection, spam filtering, rare disease discovery, and hardware fault detection. To address this issue, one popular technique is Synthetic Minority Oversampling Technique (SMOTE). SMOTE is specifically designed to tackle imbalanced datasets by generating synthetic samples for the minority class.

```python
# Importing SMOTE

from imblearn.over_sampling import SMOTE
```

```python
# Instantiate SMOTE
sm = SMOTE(random_state=27)
# Fitting SMOTE to the train set
X_smote, y_smote = sm.fit_resample(X, y) # Use fit_resample instead of fit_sample
```

```python
print('Before SMOTE oversampling X shape=',X.shape)
print('After SMOTE oversampling X shape=',X_smote.shape)
```

# Make predictions on new data

```python
# Load the pickled model
with open('best_model.pkl', 'rb') as f:
  model = pickle.load(f)


```

```python
new_data = X_test
predictions = model.predict(new_data)
```

# importing some important metrics for model evaluation

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix,f1_score,classification_report
```

```python
confusion = confusion_matrix(y_test, predictions)
sns.heatmap(confusion, annot=True, fmt='d', cmap='YlGnBu')
plt.xlabel('predictions values')
plt.ylabel('Actual values')
plt.title('confusion_matrix')
print(confusion)
```

```python
TP = confusion[1,1] # true positive 
TN = confusion[0,0] # true negatives
FP = confusion[0,1] # false positives
FN = confusion[1,0] # false negatives
```

```python
# Accuracy
print("Accuracy:-",metrics.accuracy_score(y_test, predictions))

# Sensitivity
print("Sensitivity:-",TP / float(TP+FN))

# Specificity
print("Specificity:-", TN / float(TN+FP))

# F1 score
print("F1-Score:-", f1_score(y_test, predictions))
```

```python
# classification_report
print(classification_report(y_test, predictions))
```

```python
auc=metrics.roc_auc_score(y_test,predictions)
auc
```
